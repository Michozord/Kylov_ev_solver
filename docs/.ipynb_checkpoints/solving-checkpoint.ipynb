# Solving eigenvalue problems

You can use ``krylovevsolver`` library in your Python script or in your Jupyter Notebook.

Firstly, import necessary modules.

```python
from krylovevsolver import KrylovSolver   # solver
from krylovevsolver.dff import Filter, FilterGenerator  # for filter functions

# Netgen/NGSolve imports for generating mesh and visualisation purposes
from ngsolve import *
from netgen.geom2d import SplineGeometry
from ngsolve.webgui import Draw
```

# Weak formulation of the problem
Now you have to derive weak formulation of your problem. In our example, we use eigenvalue problem of the negative Laplace operator with Neumann boundary conditions (used e.g. in wave equations). We seek eigenvalues $\omega^2$ and eigenfunctions $u$, such that 
$$-\Delta u = \omega^2 u \quad \text{ in } \Omega$$
$$\frac{\partial u}{\partial \nu} = 0 \quad\text{ on } \partial\Omega.$$
Its weak form is:
$$\int_\Omega \nabla u \cdot \nabla \varphi \, dx = \omega_h^2 \int_\Omega u \varphi \, dx.$$

Expression in the left-hand-side integral we denote with $s$, expression in the right-hand-side integral with $m$. For your problem, you have to implement this expression, using operators supported by ``netgen``/``ngsolve`` libraries. In our particular example: 

```python
from ngsolve.comp import ProxyFunction
from ngsolve.fem import CoefficientFunction

def s(u: ProxyFunction, v: ProxyFunction) -> CoefficientFunction:
    return grad(u)*grad(v)

def m(u: ProxyFunction, v: ProxyFunction) -> CoefficientFunction:
        return u*v
```

Weak formulation of the negative Laplacian eigenvalue problem is already delivered with the package in `wave_equation` module, so in this case you can just import those functions to your code.

```python
from krylovevsolver.wave_equation import s,m 
```

If you want to solve another problem, you have to reformulate it and implement functions `s` and `m` manually.



# Deriving a discrete filter function (dff)

In the next step, you have to derive appropriate **discrete filter function** (**dff**). Choose parameters `tau` (time-step), `L` (number of time-steps) and region of interest (`om_min` to `om_max`), where you seek eigenvalues. Be aware, that you choose the target interval **for square roots of eigenvalues**, not eigenvalues itself. More hints, how to choose parametres, you can find in the mathematical background section.

**Note** (**CFL condition**): square roots of all eigenvalues of your discretized FEM model *must* lie under $\omega_{\mathrm{end}} = 2/\tau$, i.e. choose $\tau < 2/\omega$. Otherwise the Krylov iteration does not converge to desired eigenvalues. 

```python
tau = 0.0056        # controlled interval up to omega_end = 2/tau = 360
L = 100
om_min, om_max = 6, 8
```

Define a `FilterGenerator` object from `krylovevsolver.dff` module with parameters `L` (number of time-steps), `tau` (time-step), `om_min` and `om_max` (target interval) and `om_end` (end of controlled interval, recommended `2/tau`). Use one of methods `chebyshev()`, `l2()` or `fourier()` to get a `Filter` object - this is your filter function (here denoted as `alpha`). Use its `plot()` method to see, whether it has desired shape - obtained dff should be higher in your target interval and possibly close 0 otherwise in whole controlled area from 0 to `om_end`.

`FilterGenerator.chebyshev()` method employs a least-squares collocation in Chebyshev nodes on $(0, \omega_{\mathrm{end}}^2)$, so you have to specify number of nodes (if result is not satisfactory, try increasing number of nodes). It also may happen, that your target interval (especially, if it is small) lies between two neighbor knots - in this case you get just a constant-zero function (very undesired!). You can plot Chebyshev nodes on your interval, using `plot_chebyshev_nodes()` method.

`FilterGenerator.l2()` method relies on a $L_2$-minimization with a quadrature rule. Optionally, you can specify number of sample points in each unit.

`FilterGenerator.fourier()` method relies on inverse Fourier transform. It is most efficient and reliable, but works for negative Laplacian eigenvalue problem only.

```python
# Use one of following:
alpha = FilterGenerator(L, tau, om_min, om_max, 2/tau).fourier() # for negative Laplacian only!
alpha = FilterGenerator(L, tau, om_min, om_max, 2/tau).chebyshev(1000)  # 1000 determines number of Chebyshev nodes for collocation
alpha = FilterGenerator(L, tau, om_min, om_max, 2/tau).l2()
```

```python
alpha.plot(0, 100)
```

This is a great example of a discrete filter function - it is significantly higher in the target interval and close to 0 otherwise.

# Defining domain and mesh

```python

geo = SplineGeometry()
p1,p2,p3,p4 = [ geo.AppendPoint(x,y) for x,y in [(0,0), (1,0),(0,1), (1,1) ] ]
geo.Append(["line", p1, p2])
geo.Append(["line", p2, p4])
geo.Append(["spline3", p4, p3, p1])

mesh = Mesh(geo.GenerateMesh(maxh=0.05))
Draw(mesh)
```

```python
solver = KrylovSolver(s, m, mesh, L, tau, alpha1, m_max = 30)
solver.discretize(order=2)
solver.solve()
solver.plot(5, 15, "")
```

```python
eigval, eigvec = solver.get_single_result(7*7, -1)
print(eigval)
solution = GridFunction(solver.fes, multidim=solver.mesh.nv)
solution.vec.data += eigvec
Draw(solution, mesh, deformation=True)
```
